```python
chart2 = pd.pivot_table(net,index = 'release_year',
               values = 'country',
               aggfunc = ['count'],
               margins = True,
               columns='type',
               margins_name = 'Total Count')
chart2
```

```python
chart2.plot(kind='line', stacked=True)
plt.xlabel('release_year')
plt.ylabel('country')
plt.title('Stacked Bar Chart for release_year and country')
plt.legend(title='Category')
plt.show()
```

```python
chart2.plot(kind='bar', stacked=True,)
plt.xlabel('release_year')
plt.ylabel('country')
plt.title('Stacked Bar Chart for release_year and country')
plt.legend(title='Category')
plt.figure(figsize=(15, 10))
plt.show()
```
